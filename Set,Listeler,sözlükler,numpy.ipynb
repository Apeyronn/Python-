## Szlk elerine erimek


u anda ilememiz gereken son ana veri tipimiz szlklerdir. Szlkler u ana kadar rendiimiz veri tiplerinden farkl olarak "gruplara" sahip olmamz salar, ancak indexe gre almak yerine Key/Value (Anahtar/Deer) iftlerini kullanlrlar. Bunun baz nemli faydalar var, ancak ncesinde nasl altn grelim.

nemli Not: Szlkler de kme parantezleriyle gsterilir, ancak bir set olarak alglanmamalar iin Key ve Value deerlerinden olumas gerekir.

```python
new_dictionary = {'key_for_this_thing': 8}
print(new_dictionary['key_for_this_thing'])
```

Szlkler, her bir key deerinin ierisinde birden fazla farkl veri tipi bulundurabilir.

```python
new_dictionary = {'key_for_a_list': [0,1,2,3,4], 'key_for_another_dictionary': {'bob': 1}}
print(new_dictionary['key_for_a_list'])
```

```python
print(new_dictionary['key_for_another_dictionary'])
print(new_dictionary['key_for_another_dictionary']['bob'])
```

```python

career_stats = {'babe_ruth': {1914: ["Red Sox", 5, 10, 0], 1915:['Red Sox', 43, 104, 4]},
                'gavvy_cravath': {1914: ['Phillies',149,604,14]}}

print(career_stats['babe_ruth'])
print(career_stats['babe_ruth'][1914])
```

```python
# Ek olarak sadece key ve value deerine de ulalabilir.
print(career_stats.keys())
print(career_stats.values())
```

Szlkleri sevmemizin nedeni, veri almada gerekten hzl olmalar ve iermek istediimiz her deere bir anahtar atamamza izin vermeleridir.
