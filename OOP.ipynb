## Pythonda durum bilgisi özniteliklerde, davranış bilgisi ise yöntemler bulunur

Bir nesne şunlardan oluşur:

**Durum:**
Nesnenin öznitelikleri tarafından temsil edilir. Ayrıca nesnenin özelliklerini yansıtır.

**Davranış:**
Nesnenin yöntemleri tarafından temsil edilir. Ayrıca nesnenin diğer nesnelere yanıtını yansıtır.

**Kimlik:**
Bir nesneye benzersiz bir isim verir ve bir nesnenin diğer nesnelerle etkileşim kurmasını sağlar.

Durum(State), davranışın ve kimliğin anlaşılması için köpek sınıfı örneğini ele alalım (yukarıda açıklandı).
Kimlik, köpeğin adı olarak düşünülebilir.
Durum veya Öznitelikler(attributes), köpeğin cinsi, yaşı veya rengi olarak düşünülebilir.
Davranış, köpeğin yemek yiyor olup olmadığı veya uyuyor olup olmadığı olarak düşünülebilir.

## Miras Yapıları(Inheritance)

Miras alma, bir sınıfın başka bir sınıftan özelliklerini türetme veya miras alma yeteneğidir. Özellikleri türeten sınıfa türetilmiş sınıf veya alt sınıf denir ve özelliklerin türetildiği sınıfa temel sınıf veya üst sınıf denir. Mirasın faydaları şunlardır:

**Gerçek dünya ilişkilerini iyi temsil eder.**

**Kodun yeniden kullanılabilirliğini sağlar. Aynı kodu tekrar tekrar yazmak zorunda kalmayız. Ayrıca, bir sınıfa daha fazla özellik eklememize izin verirken değiştirmeden önce.**

**Doğası gereği geçişlidir, yani B sınıfı bir diğer A sınıfından miras aldığında, B'nin tüm alt sınıfları otomatik olarak A sınıfından miras alır.**

### Mirasın Türleri-

**Tek Miras:**
Tek seviye mirası, türetilmiş bir sınıfın tek bir ebeveyn sınıftan özelliklerini miras almasını sağlar.

**Çoklu Seviye Miras:**
Çok seviyeli miras, bir türetilmiş sınıfın özelliklerini doğrudan bir ebeveyn sınıftan miras almasını sağlar ve bu da özellikleri kendi ebeveyn sınıfından miras alan bir ebeveyn sınıftan miras alır.

**Hiyerarşik Miras:**
Hiyerarşik seviye mirası, birden fazla türetilmiş sınıfın bir ebeveyn sınıfından özelliklerini miras almasını sağlar.

**Çoklu Miras:**
Çok seviyeli miras, bir türetilmiş sınıfın birden fazla temel sınıftan özelliklerini miras almasını sağlar.

```python
# Python code to demonstrate how parent constructors
# are called.
 
# parent class
class Person(object):
 
    # __init__ is known as the constructor
    def __init__(self, name, idnumber):
        self.name = name
        self.idnumber = idnumber
 
    def display(self):
        print(self.name)
        print(self.idnumber)
         
    def details(self):
        print("My name is {}".format(self.name))
        print("IdNumber: {}".format(self.idnumber))
     
# child class
class Employee(Person):
    def __init__(self, name, idnumber, salary, post):
        self.salary = salary
        self.post = post
 
        # invoking the __init__ of the parent class
        Person.__init__(self, name, idnumber)
         
    def details(self):
        print("My name is {}".format(self.name))
        print("IdNumber: {}".format(self.idnumber))
        print("Post: {}".format(self.post))
 
 
# creation of an object variable or an instance
a = Employee('Rahul', 886012, 200000, "Intern")
 
# calling a function of the class Person using
# its instance
a.display()
a.details()
```

```python
class Employee:
    def __init__(self, name, salary=30000):
        self.name = name
        self.salary = salary

    def give_raise(self, amount):
        self.salary += amount

        
class Manager(Employee):
    def display(self):
        print("Manager ", self.name)

    def __init__(self, name, salary=50000, project=None):
        Employee.__init__(self, name, salary)
        self.project = project

    # Add a give_raise method
    def give_raise(self, amount, bonus=1.05):
        new_amount = amount * bonus
        Employee.give_raise(self, new_amount)
    
    
mngr = Manager("Ashta Dunbar", 78500)
mngr.give_raise(1000)
print(mngr.salary)
mngr.give_raise(2000, bonus=1.03)
print(mngr.salary)
```

## Polymorphism

Polimorfizm sadece birçok formu olan anlamına gelir. Örneğin, belirli bir kuş türünün uçup uçmadığını belirlememiz gerekiyorsa, polimorfizm kullanarak bunu tek bir işlevle yapabiliriz.

Örnek: Python'da Polimorfizm

```python
class Bird:

	def intro(self):
		print("There are many types of birds.")

	def flight(self):
		print("Most of the birds can fly but some cannot.")

class sparrow(Bird):

	def flight(self):
		print("Sparrows can fly.")

class ostrich(Bird):

	def flight(self):
		print("Ostriches cannot fly.")

obj_bird = Bird()
obj_spr = sparrow()
obj_ost = ostrich()

obj_bird.intro()
obj_bird.flight()

obj_spr.intro()
obj_spr.flight()

obj_ost.intro()
obj_ost.flight()
```

Polimorfizm, aynı isimli ancak farklı işlevlere sahip olan yöntemlerin kullanımına izin veren bir kavramdır. Yani, farklı nesne türlerinin aynı yöntemi çağırması durumudur. Bu, programlamanın önemli bir kavramıdır ve kodun daha okunaklı ve yeniden kullanılabilir olmasını sağlar.

Örnek 1:

Aşağıdaki örnekte, Animal sınıfının alt sınıflarından olan Cat ve Dog sınıflarının aynı adlı make_sound() metodunu kullandığını görebilirsiniz:

```python
class Animal:
    def make_sound(self):
        pass

class Cat(Animal):
    def make_sound(self):
        print("Miyav")

class Dog(Animal):
    def make_sound(self):
        print("Hav")
```

## Kapsülleme


Kapsülleme, nesne yönelimli programlamanın (OYP) temel kavramlarından biridir. Verileri ve veriler üzerinde çalışan metodları tek bir birim içinde sarmalama fikrini açıklar. Bu doğrudan değişkenlere ve yöntemlere erişimi kısıtlar ve verilerin kazara değiştirilmesini önleyebilir. Kazara değişiklikleri önlemek için bir nesnenin değişkeni sadece nesnenin yöntemi tarafından değiştirilebilir. Bu tür değişkenlere özel değişkenler denir.
Bir sınıf, üye işlevleri, değişkenleri vb. tüm verileri kapsüllediği için bir kapsülleme örneğidir.

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.__age = age
    
    def get_age(self):
        return self.__age
    
    def set_age(self, age):
        if age >= 0:
            self.__age = age
        else:
            print("Age cannot be negative.")
```

```python
p = Person("John", 25)
p.set_age(30)
```
