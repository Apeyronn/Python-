## Miras Yaplar(Inheritance)

Miras alma, bir snfn baka bir snftan zelliklerini tretme veya miras alma yeteneidir. zellikleri treten snfa tretilmi snf veya alt snf denir ve zelliklerin tretildii snfa temel snf veya st snf denir. Mirasn faydalar unlardr:

**Gerek dnya ilikilerini iyi temsil eder.**

**Kodun yeniden kullanlabilirliini salar. Ayn kodu tekrar tekrar yazmak zorunda kalmayz. Ayrca, bir snfa daha fazla zellik eklememize izin verirken deitirmeden nce.**

**Doas gerei geilidir, yani B snf bir dier A snfndan miras aldnda, B'nin tm alt snflar otomatik olarak A snfndan miras alr.**

### Mirasn Trleri-

**Tek Miras:**
Tek seviye miras, tretilmi bir snfn tek bir ebeveyn snftan zelliklerini miras almasn salar.

**oklu Seviye Miras:**
ok seviyeli miras, bir tretilmi snfn zelliklerini dorudan bir ebeveyn snftan miras almasn salar ve bu da zellikleri kendi ebeveyn snfndan miras alan bir ebeveyn snftan miras alr.

**Hiyerarik Miras:**
Hiyerarik seviye miras, birden fazla tretilmi snfn bir ebeveyn snfndan zelliklerini miras almasn salar.

**oklu Miras:**
ok seviyeli miras, bir tretilmi snfn birden fazla temel snftan zelliklerini miras almasn salar.

```python
# Python code to demonstrate how parent constructors
# are called.
 
# parent class
class Person(object):
 
    # __init__ is known as the constructor
    def __init__(self, name, idnumber):
        self.name = name
        self.idnumber = idnumber
 
    def display(self):
        print(self.name)
        print(self.idnumber)
         
    def details(self):
        print("My name is {}".format(self.name))
        print("IdNumber: {}".format(self.idnumber))
     
# child class
class Employee(Person):
    def __init__(self, name, idnumber, salary, post):
        self.salary = salary
        self.post = post
 
        # invoking the __init__ of the parent class
        Person.__init__(self, name, idnumber)
         
    def details(self):
        print("My name is {}".format(self.name))
        print("IdNumber: {}".format(self.idnumber))
        print("Post: {}".format(self.post))
 
 
# creation of an object variable or an instance
a = Employee('Rahul', 886012, 200000, "Intern")
 
# calling a function of the class Person using
# its instance
a.display()
a.details()
```

```python
class Employee:
    def __init__(self, name, salary=30000):
        self.name = name
        self.salary = salary

    def give_raise(self, amount):
        self.salary += amount

        
class Manager(Employee):
    def display(self):
        print("Manager ", self.name)

    def __init__(self, name, salary=50000, project=None):
        Employee.__init__(self, name, salary)
        self.project = project

    # Add a give_raise method
    def give_raise(self, amount, bonus=1.05):
        new_amount = amount * bonus
        Employee.give_raise(self, new_amount)
    
    
mngr = Manager("Ashta Dunbar", 78500)
mngr.give_raise(1000)
print(mngr.salary)
mngr.give_raise(2000, bonus=1.03)
print(mngr.salary)
```

## Polymorphism

Polimorfizm sadece birok formu olan anlamna gelir. rnein, belirli bir ku trnn uup umadn belirlememiz gerekiyorsa, polimorfizm kullanarak bunu tek bir ilevle yapabiliriz.

rnek: Python'da Polimorfizm

```python
class Bird:

	def intro(self):
		print("There are many types of birds.")

	def flight(self):
		print("Most of the birds can fly but some cannot.")

class sparrow(Bird):

	def flight(self):
		print("Sparrows can fly.")

class ostrich(Bird):

	def flight(self):
		print("Ostriches cannot fly.")

obj_bird = Bird()
obj_spr = sparrow()
obj_ost = ostrich()

obj_bird.intro()
obj_bird.flight()

obj_spr.intro()
obj_spr.flight()

obj_ost.intro()
obj_ost.flight()
```

Polimorfizm, ayn isimli ancak farkl ilevlere sahip olan yntemlerin kullanmna izin veren bir kavramdr. Yani, farkl nesne trlerinin ayn yntemi armas durumudur. Bu, programlamann nemli bir kavramdr ve kodun daha okunakl ve yeniden kullanlabilir olmasn salar.

rnek 1:

Aadaki rnekte, Animal snfnn alt snflarndan olan Cat ve Dog snflarnn ayn adl make_sound() metodunu kullandn grebilirsiniz:

```python
class Animal:
    def make_sound(self):
        pass

class Cat(Animal):
    def make_sound(self):
        print("Miyav")

class Dog(Animal):
    def make_sound(self):
        print("Hav")
```

## Kapslleme


Kapslleme, nesne ynelimli programlamann (OYP) temel kavramlarndan biridir. Verileri ve veriler zerinde alan metodlar tek bir birim iinde sarmalama fikrini aklar. Bu dorudan deikenlere ve yntemlere eriimi kstlar ve verilerin kazara deitirilmesini nleyebilir. Kazara deiiklikleri nlemek iin bir nesnenin deikeni sadece nesnenin yntemi tarafndan deitirilebilir. Bu tr deikenlere zel deikenler denir.
Bir snf, ye ilevleri, deikenleri vb. tm verileri kapslledii iin bir kapslleme rneidir.

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.__age = age
    
    def get_age(self):
        return self.__age
    
    def set_age(self, age):
        if age >= 0:
            self.__age = age
        else:
            print("Age cannot be negative.")
```
